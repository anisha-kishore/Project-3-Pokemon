```python
#Imports
import pandas as pd
from pathlib import Path
```

```python
#Read csv
pokemon_stats = pd.read_csv('Pokemon_stats.csv')
pokemon_stats.head(50)
```

```python
pokemon_stats.info()
```

```python
#Put none if type 2 is empty
pokemon_stats['Type 2']=pokemon_stats['Type 2'].fillna('None')
pokemon_stats.head(5)
```

```python
#Remove duplicates
pokemon_stats.duplicated()
pokemon_stats= pokemon_stats.drop_duplicates()
```

```python
#Change M/F distinction
pokemon_stats['Name'] = pokemon_stats['Name'].str.replace('♀', ' (F)').str.replace('♂',' (M)')
```

```python
#Replace 1 with 001
pokemon_stats['Number'] = pokemon_stats['Number'].astype(str).str.zfill(3)
#pokemon_stats['Number']=pokemon_stats['Number'].astype(int)
pokemon_stats.head(50)
```

```python
pokemon_stats.info()
```

```python
#Make numbers the index
pokemon_stats.set_index('Number', inplace = True)
```

```python
pokemon_stats.head(5)
```

```python
#Export as a separate csv
pokemon_stats.to_csv('pokemon_stats_no_duplicates.csv', index=False)
```
